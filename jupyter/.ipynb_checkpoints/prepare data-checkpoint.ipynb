```python
# create the column that indecate to the number of minutes that passed from the midnight to the transaction minute
for x,i in zip(gracia['trans_date_trans_time'].values,gracia['trans_date_trans_time'].index):
    gracia.loc[i, 'minutes_from_midnight'] = int(x.split(' ')[1].split(':')[0])*60+int(x.split(' ')[1].split(':')[1])
```

```python
# create the column that indecate to the number of hours that passed from the beginning of the month to the transaction hour
for x,i in zip(gracia['trans_date_trans_time'].values,gracia['trans_date_trans_time'].index):
    gracia.loc[i, 'hours_from_month_start'] = (int(x.split(' ')[0].split('-')[2])-1)*24+int(x.split(' ')[1].split(':')[0])
```

```python
for x,i in zip(gracia['category'].values,gracia['category'].index):
    gracia.loc[i, 'category'] = x.split('_')[0]
```
