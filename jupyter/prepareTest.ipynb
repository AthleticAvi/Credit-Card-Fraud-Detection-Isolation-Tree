```python
import pandas as pd
```

```python
df=pd.read_csv('fraudTest.csv')
```

### generating data that we can use

```python
# create the column that indecate to the number of minutes that passed from the midnight to the transaction minute
for x,i in zip(df['trans_date_trans_time'].values,df['trans_date_trans_time'].index):
    df.loc[i, 'minutes_from_midnight'] = int(x.split(' ')[1].split(':')[0])*60+int(x.split(' ')[1].split(':')[1])
```

```python
# create the column that indecate to the number of hours that passed from the beginning of the month to the transaction hour
for x,i in zip(df['trans_date_trans_time'].values,df['trans_date_trans_time'].index):
    df.loc[i, 'hours_from_month_start'] = (int(x.split(' ')[0].split('-')[2])-1)*24+int(x.split(' ')[1].split(':')[0])
```

### drop data that we cant use

```python
df=df.drop(['Unnamed: 0'], axis=1)
df=df.drop(['gender'], axis=1)
df=df.drop(['street'], axis=1)
df=df.drop(['city'], axis=1)
df=df.drop(['city_pop'], axis=1)
df=df.drop(['job'], axis=1)
df=df.drop(['dob'], axis=1)
df=df.drop(['state'], axis=1)
df=df.drop(['zip'], axis=1)
df=df.drop(['lat'], axis=1)
df=df.drop(['long'], axis=1)
```

### make serial number for each transaction , the first number for the oldest transaction . for each user

```python


df["serial_num"] = ""

for cc in df['cc_num'].unique():
    indexesList=df.index[df['cc_num']==cc].tolist()
    for i,value in enumerate(indexesList):
        df['serial_num'][value] = i
```
